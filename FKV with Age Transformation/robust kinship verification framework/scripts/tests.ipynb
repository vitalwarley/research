```python
import scipy.io
import pandas as pd

# Load the .mat file
mat_file_path = '/mnt/heavy/DeepLearning/Datasets/CACD2000/celebrity2000_meta.mat'
mat_contents = scipy.io.loadmat(mat_file_path)
data = mat_contents['celebrityImageData'][0][0]
```

```python
rows = ['age', 'id', 'year_taken', 'feature', 'rank', 'lfw', 'birth', 'name']

df = pd.DataFrame()

for i in range(8):
    if i == 3:
        continue
    
    column = data[i]
    column_data = []

    for field in column:
        if i == 7:
            column_data.append(field[0][0])
        else:
            column_data.append(field[0])

    df[rows[i]] = column_data
```

```python
df.to_csv('/mnt/heavy/DeepLearning/Datasets/CACD2000/celebrity2000_meta.csv', index=False)
```

Creating dataframe with all FIW image paths 

```python
df = pd.read_csv('../rfiw2021/Track1/sample0/train_sort.txt', sep=' ', names=['id', 'f1', 'f2', 'label', '1'])
df
```
